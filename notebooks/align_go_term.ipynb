```python
import numpy as np
import pickle
import pathlib
from pathlib import Path
```

```python
ns = ["cc", "mf", "bp"]
splits = ["train", "valid", "test"]
```

```python
union_go_paths = [
    f"/data0/shaojiangyi/pprogo-flg-2/results/union_space_preds_only1/{x}/{x1}/union_go_terms.npy"
    for x in ns for x1 in splits
]
```

```python
go_emb_paths = [
    f"/data0/shaojiangyi/pprogo-flg-2/models/node_embedding/{x}_{x1}_go_name2emb.pkl"
    for x in ns for x1 in splits
]
```

```python
prefix_tuples = [(x,x1) for x in ns for x1 in splits]
```

```python
print(prefix_tuples)
```

```python
saving_dir = Path("/data0/shaojiangyi/pprogo-flg-2/models/node_embedding")
for cp, gp, (x,x1) in zip(union_go_paths, go_emb_paths, prefix_tuples):
    tmp_dir = saving_dir / x
    if not tmp_dir.exists():
        tmp_dir.mkdir(exist_ok=True)
    saving_path = tmp_dir / f"{x}_{x1}_go_emb.npy"
    go_names = np.load(cp, allow_pickle=True).tolist()
    with open(gp, "rb") as h:
        goemb = pickle.load(h)
    v = goemb[next(iter(goemb.keys()))]
    emb_matrix = np.zeros((len(go_names), v.shape[0]), dtype=np.float32)
    for i, go in enumerate(go_names):
        if (v := goemb.get(go)) is not None:
            emb_matrix[i] = v
    print(f"{x} {x1}:", emb_matrix.shape)
    np.save(saving_path, emb_matrix)
```
